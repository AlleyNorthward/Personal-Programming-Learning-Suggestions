```python
class ChineseTex(Scene):
    def construct(self):
        # 用支持中文的模板
        myTemplate = TexTemplate()
        myTemplate.add_to_preamble(r"""
            \usepackage[UTF8]{ctex}
            \setCJKmainfont{SimHei} % 这里可以换成你系统里的中文字体，比如 SimSun, Microsoft YaHei 等
        """)

        # 关键：使用 tex_template 和 tex_compiler 指定
        text = Tex(
            r"你好,Manim!这是中文渲染测试.",
            tex_template=myTemplate,
            tex_compiler="xelatex"
        )
        self.play(Write(text))
        self.wait(2)
%manim $_RF ChineseTex
```

```python
text = r"""
    \begin{tabular}{p{8 cm}}
    \hspace{2em}
    假定定要把长为\(l_1,l_2,\cdots,l_n\)的\(n\)个程序在磁带
    \(T_1\)和\(T_2\)上,且希望按照使最大检索时间取最小值的方式
    存放,即如果存放在\(T_1\)和\(T_2\)上的程序集合分别\(A\)和
    \(B\),则希望所选择的\(A\)和\(B\)使得\(max\{\sum_{i\in A}l_i,\sum_{i\in B}l_i\}\)
    取最小值.
    \end{tabular}
"""
class Test(Scene):
    def construct(self):
        my_template = TexTemplateLibrary().ctex
        tex = Tex(
            text,
            tex_template = my_template,
        )
        tex.width = config.frame_width - 1
        print(my_template.body)
        self.add(tex)
%manim $_RF Test
```
